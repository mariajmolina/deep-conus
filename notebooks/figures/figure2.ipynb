### Creation of figure 2

Imports

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import svgutils.compose as sc
from IPython.display import SVG
```

First, create figure with annotation of the architecture.

Then, overlay the annotation figure onto the svg architecture figure.

```python
fig = plt.figure(figsize=(8.,4.))

ax = fig.add_axes([0., 0., 1., 1.])

ax.set_title('Deep Learning Model Architecture', y=0.9, fontsize=12)

ax.text(-0.0, 0.6, '    Input\n  Features\n(32x32x20)', transform=ax.transAxes)
ax.text(0.15, 0.39, '  Convolutional\n       Layer\n   (32x32x32)', transform=ax.transAxes)
ax.text(0.3, 0.7, '     Max\n   Pooling\n(16x16x32)', transform=ax.transAxes)

ax.text(0.37, 0.345, '  Convolutional\n       Layer\n   (16x16x48)', transform=ax.transAxes)
ax.text(0.525, 0.61, '   Max\n Pooling\n(8x8x48)', transform=ax.transAxes)

ax.text(0.6, 0.3, 'Convolutional\n      Layer\n   (8x8x64)', transform=ax.transAxes)
ax.text(0.75, 0.525, '   Max\n Pooling\n(4x4x64)', transform=ax.transAxes)

ax.text(0.84, 0.21, 'Dense\n(1024)', transform=ax.transAxes)
ax.text(0.94, 0.425, 'Dense (32)', transform=ax.transAxes)

ax.axis('off')

ax.tick_params(axis='both', bottom=False, left=False, labelbottom=False, labelleft=False)

plt.savefig(f'/glade/work/molina/DATA/temp_figs/grl_fig2.svg', bbox_inches='tight', dpi=500, transparent=True)
plt.close()

sc.Figure("8in", "4in", 
    sc.Panel(sc.SVG("/glade/work/molina/DATA/temp_figs/convnet.svg").scale(0.8).move(320,130)),
    sc.Panel(sc.SVG("/glade/work/molina/DATA/temp_figs/grl_fig2.svg"))
    ).save("/glade/work/molina/DATA/temp_figs/compose.svg")
SVG('/glade/work/molina/DATA/temp_figs/compose.svg')
```
