### Creation of figure 2

Imports

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import svgutils.compose as sc
from IPython.display import SVG
# import local paths
from configpaths import figsave_path
```

First, create figure with annotation of the architecture.

I used the AlexNet style from https://alexlenail.me/NN-SVG/AlexNet.html and downloaded the image in SVG format.

Then, overlay the annotation figure onto the svg architecture figure.

```python
fig = plt.figure(figsize=(8.,4.))

ax = fig.add_axes([0., 0., 1., 1.])

ax.set_title('Deep Learning Model Architecture', x=0.85, y=0.925, fontsize=12)

ax.text(-0.0, 0.6, '    Input\n  Features\n(32x32x20)', fontsize=11, transform=ax.transAxes)
ax.text(0.295, 0.32, '  Convolutional\n       Layer\n   (32x32x32)', fontsize=11, transform=ax.transAxes)
ax.text(0.565, 0.765, '     Max\n   Pooling\n(16x16x32)', fontsize=11, transform=ax.transAxes)

ax.text(0.7175, 0.25, '  Convolutional\n       Layer\n   (16x16x48)', fontsize=11, transform=ax.transAxes)
ax.text(1., 0.615, '   Max\n Pooling\n(8x8x48)', fontsize=11, transform=ax.transAxes)

ax.text(1.175, 0.175, 'Convolutional\n      Layer\n   (8x8x64)', fontsize=11, transform=ax.transAxes)
ax.text(1.45, 0.475, '   Max\n Pooling\n(4x4x64)', fontsize=11, transform=ax.transAxes)

ax.text(1.615, 0.08, 'Dense\n  (32)', fontsize=11, transform=ax.transAxes)
ax.text(1.775, 0.3, 'Dense (1)', fontsize=11, transform=ax.transAxes)

ax.axis('off')

ax.tick_params(axis='both', bottom=False, left=False, labelbottom=False, labelleft=False)

plt.savefig(f'{figsave_path}/journal_fig2.svg', bbox_inches='tight', dpi=500, transparent=True)
plt.close()

sc.Figure("12in", "8in", 
    sc.Panel(sc.SVG(f"{figsave_path}/convnet.svg").scale(0.75).move(390,130)),
    sc.Panel(sc.SVG(f"{figsave_path}/journal_fig2.svg"))
    ).save(f"{figsave_path}/journal_fig2.svg")
SVG(f'{figsave_path}/journal_fig2.svg')
```
