```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.gridspec as gridspec
import matplotlib.image as mpimg
import svgutils.compose as sc
from IPython.display import SVG
```

```python
df_tp = pd.read_csv('/glade/scratch/molina/DL_proj/model25_fig6_df_tp.csv')
df_fp = pd.read_csv('/glade/scratch/molina/DL_proj/model25_fig6_df_fp.csv')
df_fn = pd.read_csv('/glade/scratch/molina/DL_proj/model25_fig6_df_fn.csv')
df_tn = pd.read_csv('/glade/scratch/molina/DL_proj/model25_fig6_df_tn.csv')
```

```python
fig = plt.figure(figsize=(13,8))

g = sns.JointGrid(x = df_tp["QVAPOR_C"],y = df_tp["SHEAR_C"], data=df_tp, xlim=(0, 11), ylim=(0, 50))

sns.scatterplot(df_tp["QVAPOR_C"], df_tp["SHEAR_C"], color='k', s=15, ax=g.ax_joint)
sns.scatterplot(df_tp["QVAPOR_F"], df_tp["SHEAR_F"], color="purple", s=15, ax=g.ax_joint)

df_scatt = pd.DataFrame({'A': df_tp["QVAPOR_C"].mean(), 
                         'B': df_tp["SHEAR_C"].mean(),
                         'C': df_tp["QVAPOR_F"].mean(), 
                         'D': df_tp["SHEAR_F"].mean(),}, index=[0])
sns.scatterplot(df_scatt["A"], df_scatt["B"], marker='*', color='k', s=500, ax=g.ax_joint, zorder=90)
sns.scatterplot(df_scatt["C"], df_scatt["D"], marker='*', color='purple', s=500, ax=g.ax_joint, zorder=90)

sns.distplot(df_tp["QVAPOR_C"], hist=False, kde=True, color="k", kde_kws={"shade": True, "linewidth":1.5}, ax=g.ax_marg_x)
sns.distplot(df_tp["QVAPOR_F"], hist=False, kde=True, color="purple", kde_kws={"shade": True, "linewidth":1.5}, ax=g.ax_marg_x)

sns.distplot(df_tp["SHEAR_C"], hist=False, kde=True, color="k", kde_kws={"shade": True, "linewidth":1.5}, ax=g.ax_marg_y, vertical=True)
sns.distplot(df_tp["SHEAR_F"], hist=False, kde=True, color="purple", kde_kws={"shade": True, "linewidth":1.5}, ax=g.ax_marg_y, vertical=True)

g.ax_joint.set_ylabel('Bulk Wind Shear (1-5 km AGL; m s$^{-1}$)', fontsize=10)
g.ax_joint.set_xlabel(' ', fontsize=10)
g.ax_joint.grid('on')
g.ax_joint.set_title("a) True Positive", pad=75., fontweight='bold')

g.ax_joint.set_yticks([0,10,20,30,40,50])
g.ax_joint.set_xticks([0,2,4,6,8,10])

g.ax_marg_y.set_ylabel('')
g.ax_marg_x.set_xlabel('')

ax1 = g.ax_marg_x
ax1.set_ylim(0, 0.4)
ax1.yaxis.set_ticks([0,0.1,0.2])
ax1.yaxis.set_ticklabels([0,0.1,0.2], fontsize=10, visible=True)
ax1.yaxis.set_ticks_position('right')

ax2 = g.ax_marg_y
ax2.set_xlim(0, 0.1)
ax2.xaxis.set_ticks([0,0.05,0.1])
ax2.xaxis.set_ticklabels([0,0.05,0.1], fontsize=10, visible=True)
ax2.xaxis.set_ticks_position('bottom')

from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='k', lw=1.5),
                Line2D([0], [0], color='purple', lw=1.5)]

g.ax_joint.legend(custom_lines, ['Current','Outlier'], fontsize=10)

plt.savefig(f"/glade/work/molina/DATA/temp_figs/journal_fig6_tp.svg", bbox_inches='tight', dpi=500, transparent=True)
plt.show()
```
